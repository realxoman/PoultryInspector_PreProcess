```python
from ultralytics import YOLO

# Load the pre-trained YOLOv8 model (you can use yolov8s.pt or larger models if needed)
model = YOLO('yolov8n.pt')

# Train the model
model.train(
    data='dataset/data.yaml',  # Path to the data.yaml file
    epochs=50,                 # Number of training epochs
    imgsz=640,                 # Input image size
    batch=16,                   # Batch size
    save=True
)
```

```python
from ultralytics import YOLO
import cv2
import os

# Load the trained model
model = YOLO('best.pt')  # Path to your trained segmentation model

# Directory for test images and output
input_dir = 'dataset/images/test'  # Path to your test images
output_dir = 'dataset/images/output'  # Path to save segmented images
os.makedirs(output_dir, exist_ok=True)

# Iterate over all test images
for image_file in os.listdir(input_dir):
    if image_file.endswith(('.jpg', '.png')):  # Process only image files
        image_path = os.path.join(input_dir, image_file)
        
        # Perform prediction with Segmentation
        results = model.predict(source=image_path, save=True)
        print(results)
```

```python
import cv2
import os

# دایرکتوری تصاویر ورودی و خروجی
input_dir = 'path/to/input/images'
output_dir = 'path/to/output/crops'

# اطمینان از وجود دایرکتوری خروجی
os.makedirs(output_dir, exist_ok=True)

# پردازش تصاویر
for result in results:
    image_path = result.path
    image = cv2.imread(image_path)
    
    # مرور bounding boxها
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # مختصات جعبه
        crop = image[y1:y2, x1:x2]  # برش تصویر
        class_id = int(box.cls[0])  # کلاس شناسایی‌شده
        
        # ذخیره تصویر برش‌خورده
        output_path = os.path.join(output_dir, f"{os.path.basename(image_path).split('.')[0]}_class{class_id}.jpg")
        cv2.imwrite(output_path, crop)
```
